# EXAONE-3.5 모델

```python
# GPU변경
```

```python
pipe=transformers.pipeline(
    task='text-generation',
    model='LGAI-EXAONE/EXAONE-3.5-2.4B-Instruct',
    device=0, # GPU 사용
    trust_remote_code=True
)
```

```python
# 메시지 템플릿: 'role'과 'content'를 키로 갖는 dict들을 아이템으로 갖는 리스트.
# role: 역할(system, user, assistant)
# content: 내용
messages=[
    {
        'role':'system',
        'content':'너는 쇼핑몰 Q&A에 올라온 질문에 답변하는 챗봇이야. \
        확정적인 답변을 하지 말고, 제품 담당자가 정확한 답변을 하기 위해 시간이 필요하다는 \
        간단하고 친절한 답변을 생성해줘.'
    },
    {
        'role':'user',
        'content':'이 다이어리에는 내년의 공휴일이 표기되어 있나요?'
    }

]
```

```python
result= pipe(messages, max_new_tokens=200)
```

```python
print(result)
# pipeline 호출 결과: 'generated_text' 키를 갖는 dict 1개를 저장하고 있는 리스트
```

pipeline은 실행할 때마다 다른 답변(텍스트)를 생성함.

pipeline의 파라미터:

* `retune_full_text`: 이전의 대화 기록을 모두 리턴할 여부. 기본값을 True.
* `so_sample`: 샘플링 전략을 사용할 지 여부를 설정. 기본값은 True.
  * `do_sample=True`: 메시지 프롬프트가 같아도 실행할 때마다 생성되는 텍스트가 달라짐.
  * `do_sample=False`
    * 메시지 프롬프트가 같으면 항상 같은 텍스트(답변)가 생성된다.
    * 가장 확률이 높은 토큰들만 선택해서 텍스트를 생성.
* 샘플링 전략
  * temperature(온도)
  * top-k 방식
  * top-p 방식    

```python
result =pipe(messages,
             max_new_tokens=200, # 답변 다양화
             return_full_text=False, # 이전답변 끄기
             do_sample=False) # 항상같은답변으로 나오게하기
print(result)
# result - [ { 'generated_text': '답변...' } ]
# 실행할 때마다 항상 같은 답변.
```

# 샘플링 전략 - temperature

```python
result= pipe(messages,
             max_new_tokens=200,
             return_full_text=False,
             temperature=10.0)
print(result)
```

```python
result= pipe(messages,
             max_new_tokens=200,
             return_full_text=False,
             temperature=0.001)
print(result)
```

temperature(온도)를 1보다 큰 값을 사용하면 문장을 생성할 때 선택되는 토큰들이 다양해 지기 때문에 갑변이 무작위짐.

temperature를 1보다 작은 값을 사용하면 높은 확률을 가진 토큰들이 선택될 가능성이 더 커짐. 생성되는 답변이 더 일관적이게 됨.

```python
# 동전 던지기
probs = [0.5, 0.5] # 앞면 확률, 뒷면 확률

np.random.multinomial(n=100,pvals=probs) # 동전 던지기 100번 실험
```

```python
# 주사위 던지기
probs =[1/6]*6
np.random.multinomial(n=600, pvals=probs)
```

logit(로짓): softmax 함수의 아규먼트. LLM에서는 어휘 사전에 포함된 각 토큰에 대한 점수.

```python
logits = [1,2,3,4,10] # 5개의 토큰의 점수
probs = scipy.special.softmax(logits)
print(probs)
```

```python
np.sum(probs)
```

```python
np.random.multinomial(n=200, pvals=probs)
```

```python
logits= np.array([1,2,3,4,100])
probs=scipy.special.softmax(logits)
print(probs)
```

```python
np.random.multinomial(n=200, pvals=probs)
```

```python
probs = scipy.special.softmax(logits / 10) # 리스트는 + * 만가능
                                           # 배열은 나누기도 가능
print(probs)
np.random.multinomial(n=200,pvals=probs)
```

```python
probs = scipy.special.softmax(logits / 100) # 리스트는 + * 만가능
                                           # 배열은 나누기도 가능
print(probs)
np.random.multinomial(n=200,pvals=probs)
```

```python
probs = scipy.special.softmax(logits / 0.1) # 리스트는 + * 만가능
                                           # 배열은 나누기도 가능
print(probs)
np.random.multinomial(n=200,pvals=probs)
```

* logit 값들을 1보다 큰 수로 나눈 후 softmax 함수로 확률을 계산하면, logit 값이 작은 토큰들의 확률이 커지는효과가 생긴다.
  * 토큰을 선택하는 다양성이 커진다.
  * 생성되는 텍스트가 다양해짐.
* logit 값들을 1보다 작은수로 나눈후에 softmax로 확률을 계산하면, logit값이 큰 토큰들의 확률을 더 크게 만드는 효과가 생김.
  * 토큰을 선택하는 다양성이 작아진다.
  * 거의 비슷한 텍스트들이 생성된다.
* LLM에서 온도(temoerature)는 logit 값들을 나눠주는 수.  

# top-k 샘플링

모델이 계산한 logit(점수) 값들을 정렬해서 최상위 k개의 토큰들 중에서 선택하는 방법.

k값이 클 수록 더 다양한 텍스트들이 생성될 수 있음.

```python
result =pipe(messages,
             max_new_tokens=200,
             return_full_text=False,
             temperature=100.5,
             top_k=100)
print(result)
```

# top-p 샘플링

모델이 출력한 logit 값으로 계산한 확률의 크기 순으로 정렬했을 때 누적 확률 p까지의 토큰을 선택하는 방법.

top_p(누적 확률) 값이 클 수록 다양한 텍스트, 작을 수록 일관된 텍스트들이 생성됨.

누적 확률 p가 동일하더라도 선택될 수 있는 토큰의 개수가 달라질 수 있음.

top-p 샘플링은 다양한 확률 분포에 유연하게 대처할 수 있음.

top-k 샘플링은 선택하는 토큰들의 개수를 고정하기 때문에, 높은 확률을 가진 토큰이 선택에서 제외될 가능성이 있음.

```python
result=pipe(messages,
            max_new_tokens=200,
            return_full_text=False,
            temperature=1.5,
            top_p=0.9)
print(result)
```

```python
result=pipe(messages,
            max_new_tokens=200,
            return_full_text=False,
            temperature=1.5,
            top_p=0.3)
print(result)
```
